```python
import pandas as pd
```

```python
file_1= pd.read_csv('Churn_Modelling_AccDetails.csv')
```

```python
file_2= pd.read_csv('Churn_Modelling_CustDetails.csv')
```

```python
data_1=pd.DataFrame(file_1)
```

```python
data_1.head()
```

```python
data_1.info()
```

```python
data_1.CustomerId.nunique()
```

```python
data_2=pd.DataFrame(file_2)
```

```python
data_2.head()
```

```python
data_2.info()
```

```python
data_2.CustomerId.nunique()
```

```python
df = pd.merge(data_2, data_1, on='CustomerId', how='outer')
df.head()
```

```python
df.set_index('CustomerId', inplace=True)
df.head()
```

```python
df.info()
```

```python
df.describe()
```

## Surname

```python
df[pd.isnull(df['Surname'])]
```

```python
df.Surname.fillna('noname' , inplace=True)
```

## Credit Score

```python
df[pd.isnull(df['CreditScore'])]
```

```python
a = df.CreditScore.mean()
a
```

```python
df.CreditScore.fillna(int(a) , inplace=True)
```

## Geography

```python
df[pd.isnull(df['Geography'])]
```

```python
df.Geography.value_counts()
```

```python
df.groupby(['Geography','Gender']).median()
```

```python
df.groupby(['Geography','Tenure']).median()
```

```python
df.loc[df.Balance > 100000, 'Geography'] = df.loc[df.Balance > 100000, 'Geography'].fillna('Germany')
```

```python
df.loc[(df.Balance == 0) & (df.Tenure == 10), 'Geography'] = df.loc[(df.Balance == 0) & (df.Tenure == 10), 'Geography'].fillna('Spain')
```

```python
df.Geography.fillna('France' , inplace=True)
```

```python
df[pd.isnull(df['Geography'])]
```

```python
df.loc[df.Geography == 'France', 'Gender'] = df.loc[df.Geography == 'France', 'Gender'].fillna('Female')
```

```python
df[pd.isnull(df['Gender'])]
```
